```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from fbprophet import Prophet
from fbprophet import __version__ as prophet_version
print(prophet_version)
```

covid data from: https://health-infobase.canada.ca/covid-19/epidemiological-summary-covid-19-cases.html?stat=num&measure=active#a2

# Load COVID data

```python
covid_data = pd.read_csv('covid19-download.csv')
covid_data.head()
```

# Pick right columns

```python
covid_data.columns
```

```python
covid_canada_data = covid_data.loc[covid_data['prname'] == 'Canada'][['date', 'numtoday']]
covid_canada_data = covid_canada_data.rename({'date': 'ds', 'numtoday': 'y'}, axis=1)
covid_canada_data['ds'] = pd.to_datetime(covid_canada_data['ds'])
covid_canada_data.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(covid_canada_data['ds'], covid_canada_data['y'])
plt.show()
```

# Looks like 3 change points should be good

```python
model = Prophet().fit(covid_canada_data)
```

```python
future_dataframe = model.make_future_dataframe(periods=90, include_history=True)
```

```python
future_predict = model.predict(future_dataframe)
#future_predict = future_predict[['ds', 'yhat']]
```

```python
model.plot_components(future_predict);
```

```python
plt.figure(figsize=(15, 5))
plt.plot(future_predict['ds'], future_predict['yhat'], c="blue")
plt.plot(future_predict['ds'], future_predict['yhat_upper'], c="teal")
plt.plot(future_predict['ds'], future_predict['yhat_lower'], c="teal")
plt.fill_between(future_predict['ds'], 
                 future_predict['yhat_upper'], 
                 future_predict['yhat_lower'],
                 color="teal",
                 alpha=0.5)
plt.plot(covid_canada_data['ds'], covid_canada_data['y'], c="red")
plt.show()
```

## Try with log data

```python
covid_canada_data = covid_data.loc[covid_data['prname'] == 'Canada'][['date', 'numtoday']]
covid_canada_data = covid_canada_data.rename({'date': 'ds', 'numtoday': 'y'}, axis=1)
covid_canada_data['ds'] = pd.to_datetime(covid_canada_data['ds'])
covid_canada_data['y'] = np.log(covid_canada_data['y']+1)
covid_canada_data.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(covid_canada_data['ds'], covid_canada_data['y'])
plt.show()
```

# Looks like 3 change points should be good

```python
model = Prophet(interval_width=0.5).fit(covid_canada_data)
```

```python
future_dataframe = model.make_future_dataframe(periods=30, include_history=True)
```

```python
future_predict = model.predict(future_dataframe)
#future_predict = future_predict[['ds', 'yhat']]
```

```python
model.plot_components(future_predict);
```

```python
plt.figure(figsize=(15, 5))
plt.plot(future_predict['ds'], np.exp(future_predict['yhat'])-1, c="blue")
plt.plot(future_predict['ds'], np.exp(future_predict['yhat_upper'])-1, c="teal")
plt.plot(future_predict['ds'], np.exp(future_predict['yhat_lower'])-1, c="teal")
plt.fill_between(future_predict['ds'], 
                 np.exp(future_predict['yhat_upper'])-1, 
                 np.exp(future_predict['yhat_lower'])-1,
                 color="teal",
                 alpha=0.5)
plt.plot(covid_canada_data['ds'], np.exp(covid_canada_data['y'])-1, c="red")
plt.show()
```
